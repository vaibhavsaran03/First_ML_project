```python
HDF=HDF.drop('X1 transaction date',axis=1)
```

```python
HDF.drop('X2 house age',axis=1)
```

```python
x=HDF[['X3 distance to the nearest MRT station',
       'X4 number of convenience stores', 'X5 latitude', 'X6 longitude']]
y=HDF['Y house price of unit area']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.60, random_state=50)
```

```python
x_train
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lm=LinearRegression()
```

```python
lm.fit(x_train,y_train)
```

```python
coeff_df=pd.DataFrame(lm.coef_,x.columns,columns=['Coeffecient'])
```

```python
coeff_df
```

```python
prediction=lm.predict(x_test)
```

```python
plt.scatter(y_test,prediction)
```
