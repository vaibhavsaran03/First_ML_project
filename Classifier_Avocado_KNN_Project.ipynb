```python
import numpy as np
import pandas as pd
import sklearn as sl
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
avo=pd.read_csv("avocado.csv")
```

```python
avoDF=avo
```

```python
avoDF
```

```python
avoDF=avoDF.drop("Unnamed: 0",axis=1)
```

```python
avoDF.isna().sum()
```

```python
avoDF['type'].value_counts()
```

```python
from sklearn import preprocessing
```

```python
le = preprocessing.LabelEncoder()
```

```python
avoDF['type']= le.fit_transform(avoDF['type'])
```

```python
avoDF['type'].unique()
```

```python
avoDF
```

```python
avoDF['region'].value_counts()
```

```python
cities = pd.get_dummies( avoDF['region'])
```

```python
cities
```

```python
avoDF = pd.concat([avoDF, cities], axis=1)
```

```python
avoDF
```

```python
avoDF.columns
```

```python
avoDF=avoDF.drop('region',axis=1)
```

```python
avoDF.corr().head(15)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x=avoDF.drop(columns=['Date','type','year'],axis=1)
y=avoDF['type']
```

```python
x_train,x_test,y_train,y_test =train_test_split(x,y,test_size=0.3,random_state=2)
```

```python
x_train
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(n_neighbors=4)
```

```python
knn.fit(x_train,y_train)
```

```python
predictions=knn.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error
```

```python
mean_squared_error(y_test,predictions)
```

```python
knn.score(x_train,y_train)
```
